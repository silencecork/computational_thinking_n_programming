## **Python函數類型**
- 在Python中有兩種不同的函數
  - **內建函數** (Built-in Functions)：Python預設提供
    - e.g. print(), type(), float(), int()
  - **自訂函數** (Function that we define ourselves)：我們自己寫的並且之後拿來使用
  - 我們把**內建函數視為保留字**，也就是其他變數的命名不能夠使用 

## **定義函數**
- 在Python中，函數是一些可重用的程式碼，可以定義輸入的引數(arguments)，在程式碼中進行一些計算，然後從函數中回傳一個或多個結果
- 定義函數時，使用保留字**def**
- 使用函數，成為呼叫(call)或是調用(invoke)
- 呼叫函數是使用函數的名稱，並加上括號結尾，在括號中可寫下要傳入到函數中的引數(argument)

### 型態轉換 Type Conversion
- 當您將整數和浮點數放在表達式中時，該整數將偷偷的被轉換為浮點數
- 您可以使用內建函數int()和float()很明確的進行轉換

````python
print(float(99) / 100)

i = 42
print(type(i))

f = float(i)
print(f)

print(type(f))

print(1 + 2 * float(3) / 4 - 5)
````

### 字串轉換 String Conversion
- 您也可以使用int()或float()，讓字串轉換成數字
- 當字串無法轉換成數字時，例如字串的內容不是數字，程式就會出現錯誤(error)

````python
sval = '123'
print(type(sval))

print(sval + 1)
````

````python
sval = '123'
ival = int(sval)
print(type(ival))
print(ival + 1)
````

## **自訂函數**

### 建立自訂函數 Building our Own Functions
- 我們創建一個自訂的新函數，使用**def**關鍵字及括號，括號中的可以擺放引數 (argument)
- 函數內的程式碼，我們進行縮排(Indent)，表明程式碼是屬於函數的
- 這只是定義了一個新的函數，但在你呼叫他之前，他不會執行


````python
# 定義函數
def print_lyrics():
    print("I am a lumberjack, and I'm okay.")
    print("I sleep all night and I work all day.")
````

#### 函數在我們呼叫之前並不會正常執行

````python
x = 5
print('Hello')

# 定義函數
def print_lyrics():
    print("I am a lumberjack, and I'm okay.")
    print("I sleep all night and I work all day.")

print('Yo')
x = x + 2
print(x)
````

#### 呼叫定義好的函數
- 當我們定義好一個函數後，我們可以呼叫(call)或調用(invoke)函數，想呼叫幾次就幾次
- 這就是儲存並重複使用的模式 (**store and reuse pattern**)

````python
x = 5
print('Hello')

# 定義函數
def print_lyrics():
    print("I am a lumberjack, and I'm okay.")
    print("I sleep all night and I work all day.")

print('Yo')

# 呼叫函數
print_lyrics()

x = x + 2
print(x)
````

### 引數 Arguments
- 引數是我們在呼叫(call)/調用(invoke)函數時，傳遞給函數當作輸入的數值
- 我們可以在每次使用函數時，傳入不同的引數，讓函數去執行不同的工作
- 我們將引數放在函數名稱後的括號中
```
big = max('Hello World')
```
  - 上面的程式碼，'Hello World'就是引數Argument

### 參數 Parameters
- 參數是我們在函數定義中使用的變數，它允許函數中的程式碼訪問傳遞到函數的引數(argument)

````python
def greet(lang):
    # lang是參數parameter
    if lang == 'es':
        print('Hola')
    elif lang == 'fr':
        print('Bonjour')
    else:
        print('Hello')

# 'en'是引數 argument
greet('en')

# 'es'是引數 argument
greet('es')

# 'fr'是引數 argument
greet('fr')
````

### 回傳值1 Return Value
- 函數通常會接受其參數，進行一些計算並返回一個值
- 這個值就是函數運算的結果，回到原本呼叫(call)/調用(invoke)函數的程式碼中
- return關鍵字就是用於此目的


````python
def greet():
    return 'Hello'

print(greet(), 'Glenn')
print(greet(), 'Sally')
````

### 回傳值2 Return Value
- 一個有「產出(fruitfull)」的函數是會產生結果(或返回值)的函數
- return便結束了函數的執行，並回傳函數運作的結果

````python
def greet(lang):
    if lang == 'es':
        return 'Hola'
    elif lang == 'fr':
        return 'Bonjour'
    else:
        return 'Hello'

print(greet('en'), 'Glenn')
print(greet('es'), 'Sally')
print(greet('fr'), 'Michael')
````

### 多引數/參數 Multiple Arguments/Parameters
- 我們可以在函數定義中定義多個參數
- 呼叫/調用函數時，我們只需添加更多引數
- 我們要讓呼叫函數時，傳遞的引數與函數的參數數量與順序相同

````python
# 定義了兩個參數
def addtwo(a, b):
    added = a + b
    return added

# 呼叫/調用函數時，引數的順序和數量要與函數的參數相同
x = addtwo(3, 5)
print(x)
````

## **無產出函數 Void Function**
- 當一個函數沒有回傳值時，我們稱為Void Function
- 有回傳值的函數稱為有產出(fruitful)函數
- 沒有回傳值的函數是無產出(not fruitful)函數


## **要轉成函數/不要轉成函數**
- 將您的程式碼組織為一段一段(paragraphs)的，以便釐清完整想法並為其命名
- 別重複做同樣的事情，將程式碼處理好並能正常執行後重複使用它
- 如果有些程式碼的邏輯太長或太複雜，將程式碼的邏輯切分成較小的區段，每個區段獨立成為一個函數
- 可以為常常重複使用的程式碼建立資料庫，也許還可以與您的朋友分享
